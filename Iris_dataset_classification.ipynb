# Reading and displaying data

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

iris = load_iris()
print(iris.data)
print(iris.target)
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target_names[iris.target]
df.head()
```

```python
df.info() #Our dataset doesn't have any Nan so we don't have to do much preprocessing
```

```python
df.describe().T
```

```python
print(df.isnull().sum()) # we can check to be sure if there's no Nan
```

# Building a model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from mlxtend.plotting import plot_confusion_matrix
```

```python
X = df[['petal length (cm)','petal length (cm)']]
y=df['target']
```

Working with LogisticRegression is better to standarize our features

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

acc = accuracy_score(y_test, y_pred)
print(f'Accuracy: {acc}')
```

```python
plot_confusion_matrix(confusion_matrix(y_test, y_pred), figsize=(5,5))
plt.show()
```

We want to maximize values on the diagonal and we se that there is one wrong prediction: model predicted 2 but it was 1

```python
print(classification_report(y_test, y_pred, target_names = iris.target_names))
```
